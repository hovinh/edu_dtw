```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# An example of Control Chart

# Data preparation
np.random.seed(36)
x = np.arange(start=1, stop=21, step=1)
y = np.random.normal(loc=0, scale=1, size=20)
mean, sigma = 0, 1
ucl = mean + sigma
lcl = mean - sigma
outliers = [(idx, val) for idx, val in zip(x, y) if ((val < lcl) or (val > ucl))] 


# Plot
plt.figure(figsize=(14,8))

plt.plot(x, y, linestyle='-', marker='o', color='black', alpha=1)
plt.scatter([i[0] for i in outliers], [i[1] for i in outliers], c='red', s=100, alpha=0.7)#marker='o', color='red')

font_dict = {"family": "serif", "color": "blue", "size": 10}
plt.axhline(mean, color='blue')
plt.text(19, mean + 0.1, s=f"mean={mean:.3f}", fontdict=font_dict)

font_dict = {"family": "serif", "color": "red", "size": 10}
plt.axhline(ucl, color = 'red', linestyle = 'dashed')
plt.text(19, ucl + 0.1, s=f"UCL={ucl:.3f}", fontdict=font_dict)
plt.axhline(lcl, color = 'red', linestyle = 'dashed')
plt.text(19, lcl - 0.2, s=f"LCL={lcl:.3f}", fontdict=font_dict)

plt.title('Univariate Control Chart')
plt.xlabel('Timestamp')
plt.xticks(x, x)
plt.ylabel('Sensor')
plt.show()
```
